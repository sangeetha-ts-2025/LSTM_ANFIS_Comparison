```python
import myanfis
import os
from tensorflow.keras.utils import plot_model
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sys
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf

df = pd.read_csv('3kmhr_sub16_emg_grf_knee.csv')
df2 = df.copy()

minmaxScaler = MinMaxScaler()
df2['emg'] = minmaxScaler.fit_transform(df2[['emg']])
df2['grf'] = minmaxScaler.fit_transform(df2[['grf']])
df2['knee'] = minmaxScaler.fit_transform(df2[['knee']])
df2['ankle'] = minmaxScaler.fit_transform(df2[['ankle']])

X = df2.iloc[:6144, [0, 1, 2]]
Y = df2.iloc[:6144, -1]

param = myanfis.fis_parameters(
    n_input=3,
    n_memb=4,
    batch_size=32,
    memb_func='gaussian',
    optimizer='sgd',
    loss=tf.keras.losses.MeanAbsoluteError(),
    n_epochs=450
)

kfold = KFold(n_splits=2)
histories = []

for train_index, test_index in kfold.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
    
    fis = myanfis.ANFIS(n_input=param.n_input,
                         n_memb=param.n_memb,
                         batch_size=param.batch_size,
                         memb_func=param.memb_func,
                         name='firstAnfis')
    
    fis.model.summary()
    
    fis.model.compile(optimizer=param.optimizer,
                      loss=param.loss,
                      metrics=['mae'])
    
    history = fis.fit(X_train, Y_train,
                      epochs=param.n_epochs,
                      batch_size=param.batch_size,
                      validation_data=(X_test, Y_test))
    
    histories.append(history)

print("predicting..............................................")
predicted_output = fis.model.predict(X_test)
output = minmaxScaler.inverse_transform(predicted_output)

output_df2 = pd.DataFrame(output)
output_df2.to_csv('3kmhr_sub16_emg_grf_knee_output.csv', index=False)

print("printing the inverse transform of predicted output")
print(type(output))
print(output)
original_data = np.exp(output)

print("printing the original data without exponent")
print(original_data)

fis.plotmfs()
fis.plotmfs(show_initial_weights=True)

print("the member functions are here--------------")
pd.DataFrame(histories[0].history).plot()
pd.DataFrame(histories[1].history).plot()
plt.show()
```
