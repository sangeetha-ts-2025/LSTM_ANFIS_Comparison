```python
# To check the performance Membership function is rediced to:4 when compared to 8 in Mustafa_version5
import myanfis
import os
from tensorflow.keras.utils import plot_model
# Veri işlemeri için kullanılacak olan kütüphane
import matplotlib.pyplot as plt
import pandas as pd

# sayısal işlemler için kullanılan kütüphane
import numpy as np

# Model için gerekli bir kütüphane
import sys
sys.maxsize

# KFold 
from sklearn.model_selection import KFold

# Scaler
from sklearn.preprocessing import MinMaxScaler

# Model için gerekli olan makine öğrenmesi kütüphanesi
import tensorflow as tf
df = pd.read_csv('moment_1kmhr_sub1_emg_grf_ankle_data_used_1.csv')
df.head()
df.describe()
minmaxScaler = MinMaxScaler()

# verimizin bir adet kopyasını aldık ki yaptığımız değişiklikler df'i etkilemesin
df2 = df.copy()

# aşağıdaki kısımdaysa verilerimizi 0-1 arasına yerleştiriyoruz
df2['emg'] = minmaxScaler.fit_transform(df2[['emg']])
df2['grf'] = minmaxScaler.fit_transform(df2[['grf']])
df2['ankle_moment'] = minmaxScaler.fit_transform(df2[['ankle_moment']])
#X = df2.iloc[:1024,[0,1]]## first 1024 rows are used; 512 for training and 512 rows for testing 
#Y = df2.iloc[:1024,-1]## 41 is taken because we selected 960 rows(which is multiple of 32)1001-41=960
X = df2.iloc[:6144,[0,1]]
Y = df2.iloc[:6144,-1]
print(X)
print(Y)
#x = df2.iloc[:-100,[0,1]]
#y = df2.iloc[:-100,-1]
param = myanfis.fis_parameters(
        n_input = 2,                # no. of Regressors
        n_memb =4,                 # no. of fuzzy memberships
        batch_size =32,            # 16 / 32 / 64 / ...
        memb_func = 'gaussian',      # 'gaussian' / 'gbellmf' / 'sigmoid'
        optimizer = 'sgd',          # sgd / adam / ...
        loss = tf.keras.losses.MeanAbsoluteError(),               # mse / mae / huber_loss / mean_absolute_percentage_error / ...
        n_epochs = 400               # 10 / 25 / 50 / 100 / ...
        )
kfold = KFold(n_splits=2)
histories = []

# Fold'larımız için gerekli indexleri kfold.split ile alarak X_train, X_test, Y_train, Y_test olarak aldık
# ve Fold Fold eğitimini gerçekleştirdik
for train_index, test_index in kfold.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]
    
    #print(X_train.shape)
    #print(X_test.shape)
    #print(Y_train.shape)
    #print(Y_test.shape)
    # Tensorflow.Keras modellerinden üretilmiş ANFIS'i gerekli parametreler ile çağırdık
    fis = myanfis.ANFIS(n_input = param.n_input,
                    n_memb = param.n_memb,
                    batch_size = param.batch_size,
                    memb_func = param.memb_func,
                    name = 'firstAnfis' # buradaki ismi değiştirin
                    )
    fis.model.summary()
    #plot_model(fis.model, to_file, show_shapes=True)
    #img = plt.imread(to_file)
    #plt.imshow(img)
    #plt.axis('off')
    #plt.show()
    # modelimizi aşağıdaki parametreler ile derledik
    fis.model.compile(optimizer=param.optimizer,
                      loss=param.loss,
                      metrics=['mae']  # ['mae', 'mse']
                      )
    # burada modelimizin eğitimini başlattık
    # daha sonrasında modelin eğitim sonuçlarını history'e atama yaptık 
    history = fis.fit(X_train, Y_train,
                  epochs=param.n_epochs,
                  batch_size=param.batch_size,
                  validation_data = (X_test, Y_test),
                  )
    # model fold fold eğitim yapacağı için eğitim sonuçlarını histories adlı listede tuttuk
    histories.append(history)
#plot_model(fis.model, to_file='anfis_model.png', show_shapes=True)
print("predicting..............................................")
fis.model.fit(X_test, Y_test, epochs=400)
print(X_test)
print(Y_test)
predicted_output=fis.model.predict(X_test)
print(f"Predicted Output  is here............: {predicted_output}")

#------------ wRITING THE NORMALIZED FORM OF PREDICTED OUTPUT, Convert the dictionary to a Pandas DataFrame-------
#output_df1 = pd.DataFrame(predicted_output)

# Save the DataFrame to a CSV file
#output_df1.to_csv('predicted_output1.csv', index=False)


print("End of predicted output")
output=minmaxScaler.inverse_transform(predicted_output)

#------------WRITING THE  PREDICTED OUTPUT(NOT NORMALIZED FORM)------------
# Convert the dictionary to a Pandas DataFrame
output_df2 = pd.DataFrame(output)

# Save the DataFrame to a CSV file
output_df2.to_csv('moment_1kmhr_sub1_emg_grf_ankle_output.csv', index=False)

print("printing the inverse transform of predicted output")
print(type(output))
print(output)

original_data = np.exp(output)

# Print the result
print("printing the original data without exponent")
print(original_data)
fis.plotmfs()
fis.plotmfs(show_initial_weights=True)
#fis.plot_output_mf()    
#sample_input = np.array([[14.67101541]])  # Example input for testing
#predicted_output = fis.model.predict(sample_input)
#print(f"Predicted Output: {predicted_output}")

print("the member functions are here--------------")
pd.DataFrame(histories[0].history).plot()
pd.DataFrame(histories[1].history).plot()
#plt.show() notworking
```
